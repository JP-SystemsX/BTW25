# AutoGluon (AG)
- A very Simple but well working AutoML Approach
- It tests out several common Forecasting approaches in a given time
- Combines the best into a weighted Ensemble

```python
import pandas as pd
from autogluon.timeseries import TimeSeriesDataFrame, TimeSeriesPredictor
```

```python
df = pd.read_csv("data/european_wholesale_electricity_price_data_hourly/european_wholesale_electricity_price_data_hourly/Germany.csv")
df = df.sort_values(by="Datetime (UTC)")
test = df[-24:]  # The last day is test data (for visualization only real test set should be much larger)
train = df[:-24]
train.head(5)
```

```python
# Convert Data Into The Format requested by AG
train_data = TimeSeriesDataFrame.from_data_frame(
    train,
    id_column="Country",
    timestamp_column="Datetime (UTC)",
)
test_data = TimeSeriesDataFrame.from_data_frame(
    test,
    id_column="Country",
    timestamp_column="Datetime (UTC)",
)
train_data.head()
```

```python
# Setup & Train AG
predictor = TimeSeriesPredictor(
    target="Price (EUR/MWhe)",  # Which Column shall be Forecasted
    prediction_length=24,  # 24h forecast
    freq="h",  # one forecast per hour
)
predictor = predictor.fit(
    train_data=train_data,
    time_limit=10*60,  # 10 min (in seconds) - time for training (more time sometimes leads to better results)
    presets="medium_quality",  # Sets the remaining Hyperparameters to good defaults during development (for deployment use "best_quality" it takes longer but is usually better)
)
```

```python
# Plot Predictions
predictions = predictor.predict(data=train_data)
predictor.plot(pd.concat([train_data[-48:], test_data]), predictions=predictions, max_num_item_ids=1)
```

# Further Work Ideas
## Optimizing & Tuning 
- AG supports several approaches to tune it to squeeze the last bit of performance out of it
#### Hyperparameter Tuning
- AG supports [Hyperparameter-tuning](https://auto.gluon.ai/stable/tutorials/timeseries/forecasting-indepth.html#hyperparameter-tuning) 
- Will likely take much longer to find solution but might be much better
#### Include More Data
- AG Supports Known and Unknown Covariates 
    - Covariates ~ Time-Series that correlate to the Time-Series that shall be predicted and provide so information to support predicting our target, e.g. Weather and Solar Energy Production (On sunny days more is produced than on Cloudy Days) 
    - Known-Covariates ~ Covariates that are known for the future e.g. It is known which week-day is tomorrow, or when the next public Holiday will be
    - Unknown-Covariates ~ Covariates we only know in hindsight, e.g. The Weather we know how it was yesterday and today, but we don't know how it will be tomorrow (Though we could make predictions)
- AG also supports Static Features, the Intuition is:
    - Those Features make it easier to find Patterns across different Time-Series, by helping the model to cluster or rather compare the Time-Series via those features
    - E.g. Time-Series 1 & 2 Belong to Germany and Time-Series 3 to France --> Shall the model now Forecast a 4th Time-Series which belongs to Germany then it "knows" that patterns found on 1 & 2 are likely more relevant than those specific to 3  
- [Finetune Large Chronos](https://auto.gluon.ai/stable/tutorials/timeseries/forecasting-chronos.html) (requires GPU) 
    - Note: Chronos has also it's own path by us, because Chronos is a very powerful approach and using it inside of AG limits your opportunities for improvement and customization (though it makes it much easier to use)
- AG promises that it can generalize Across several similar Time-Series, so including more Data, e.g. from different Nations, might help too
